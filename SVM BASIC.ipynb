#**S V M** --SUPPORT VECTOR MACHINE ALGORITHM

Support Vector Machine or SVM is one of the most popular Supervised Learning Algorithm, which is used for Classification as well as Regression Problems.



The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space into classes so it is easy put the new data point in the correct category in the future This best decision boundary is called a hyperplane

SVMs use a boundary called a hyperplane to partition data into groups of similar class values in two and three dimensions If the data points can be separated perfectly by the straight line or flat surface, they are said to be linearly separable



image.png

In two dimensions, the task of the SVM algorithm is to identify a line that separates the two classes.



There is more than one choice of dividing line between the groups of circles and squares. Three such possibilities are labeled a, b, and c.



**MAXIMUM MARGIN HYPERPLANE [MMI]**

SVM algorithm will choose Maximum Margin Hyperplane (MMI) that creates the greatest separation between the two classes

Although any of the three lines separating the circles and squares would correctly

classify all the data points, it is likely that the line that leads to the greatest separation will generalize the best to the future data

• The maximum margin will improve the chance that, in spite of random noise, the points will remam on the correct side of the boundary

**SUPPORT VECTORS**

Are the points from each class that are the closest to the MMH; each class must have the at least one support vector,but it is possible to have more than one.

using the support vectors alone,it is possible to define the MMH . This is a key feature of SVMS;the support vectors provide a very compact way to store a classification model , even if the no.of features is extremely large
